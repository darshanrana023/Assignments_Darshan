# Implement a basic linear regression model.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

df=load_diabetes()
```

```python
x=df.data[:,np.newaxis,2]
y=df.target
```

```python
# Split the dataset into training and testing sets
x_train , x_test, y_train , y_test= train_test_split(x,y,test_size=0.2)
```

```python
# Create a Linear Regression mode
model=LinearRegression()
```

```python
#Train the model on training data
model.fit(x_train,y_train)
```

```python
#make prediction
y_pred=model.predict(x_test)
```

```python
r2 = r2_score(y_test, y_pred)
```

```python
print(f"R-squared value: {r2}")
```

```python
plt.scatter(x_test,y_test,color='Blue')
plt.plot(x_test,y_pred,color='Red')
plt.title('Linear Regression plot')
plt.xlabel('BMI')
plt.ylabel('Disease Progression')
plt.grid(True)
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

# Load the dataset
# Replace 'your_dataset.csv' with your actual dataset file path or URL
data = pd.read_csv('Admission.csv')

# Allow users to select target and predictor variables
target_variable = input("Enter the target variable: ")
predictor_variables = input("Enter predictor variables (comma-separated): ").split(',')

# Prepare the data
X = data[predictor_variables]
y = data[target_variable]

# Train a linear regression model
model = LinearRegression()
model.fit(X, y)

# Display regression coefficients
print("Regression Coefficients:")
for variable, coefficient in zip(predictor_variables, model.coef_):
    print(f"{variable}: {coefficient}")

# Make predictions on the same dataset
y_pred = model.predict(X)

# Calculate R-squared value
r2 = r2_score(y, y_pred)
print(f"R-squared value: {r2}")
```
